```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn  as sns
```

```python
iris = pd.read_csv("Iris.csv")
iris
```

```python
sns.pairplot(iris) # hue='variety'
```

```python
titanic = pd.read_excel('titanic3.xls')

# Visualisation des categories
sns.catplot(x='pclass', y='age', data=titanic, hue='sex')
```

```python
sns.boxplot(x='pclass', y='age', data=titanic, hue='sex')
```

```python
# Visualisation des distriburion
sns.distplot(titanic['fare'])
```

```python
sns.jointplot('age', 'fare', data=titanic, kind='kde') # kde=kernel density estimation
```

```python
# Fonction heatmap utile pour afficher des matrices de correlation
sns.heatmap(titanic.corr())
```
