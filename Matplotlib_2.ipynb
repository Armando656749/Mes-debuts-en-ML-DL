# 1. (5e) Graphique de classification avec plt.scatter

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

x = iris.data
y = iris.target
names = list(iris.target_names)

print(f'x contient {x.shape[0]} exemples et {x.shape[1]} variables')
print(f' il y a {np.unique(y).size} classes')
```

```python
plt.scatter(x[:,0], x[:,1], c=y) # c=y effectue une colorration selon la nature de la classe
plt.xlabel("longueur sepal")
plt.ylabel("largeur sepal")
```

# 2. (4e) Graphique 3D

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
ax = plt.axes(projection="3d") # lorsqu'on cree des graphiques 3D on est oblige d'utiliser la programmation oriente objet
ax.scatter(x[:,0], x[:,1], x[:,2], c=y)
```

```python
# Visualisation des surface avec matplolib
f = lambda x,y:np.sin(x)+np.cos(x+y)
```

```python
X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)

X, Y = np.meshgrid(X, Y)
Z = f(X, Y)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma')
```

# 3. (3e) Histogramme

```python
plt.hist(x[:,0]) # par defaut bins=10
```

```python
plt.hist(x[:,0], bins=20)
plt.hist(x[:,1], bins=20)
```

```python
# On peut aussi visualiser la distribution des donnees lorsqu'elle suit deux variables
plt.hist2d(x[:,0], x[:,1], cmap='Blues')
plt.xlabel('longueur sepal')
plt.ylabel('largeur sepal')
plt.colorbar()
```

```python
from scipy import misc
face = misc.face(gray=True)
#plt.imshow(face, cmap='gray')
plt.hist(face.ravel(), bins=255)
plt.show()
```
