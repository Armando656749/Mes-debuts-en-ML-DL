```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2, 10)
y = x**2
```

```python
plt.plot(x,y, label="x^2", lw=3, ls="--", c='blue')
```

# Le cycle de vie d'une figure

```python
plt.figure(figsize=(12,8)) # la fenetre de travail et les parametre sont en inch pas en cm
plt.plot(x, y, label="x^2 ou quadratique")
plt.plot(x, x**3, label="x^3 ou cubique")
plt.title("figure1")
plt.xlabel("axe x")
plt.ylabel("axe y")
plt.legend()
plt.show()
# pour sauvegarder la figure, tu peux faire plt.savefig("text.png")
```

# Subplot

```python
plt.subplot(2,1,1)
plt.plot(x, y, c='red')
plt.subplot(2,1,2)
plt.plot(x, y, c='blue')
```

# Methode oriente objet pour faire des graphes

```python
fig, ax = plt.subplots(2, 1, sharex=True)
ax[0].plot(x, y)
ax[1].plot(x, np.sin(x))
plt.show()
```

# Exercice

```python
dataset = {f"experience{i}": np.random.randn(100) for i in range(4)}
dataset
```

```python
def graphique(dataset):
    n = len(dataset)
    plt.figure(figsize=(12,8))
    for k, i in zip(dataset.keys(), range(1,n+1)):
        plt.subplot(n,1, i)
        plt.plot(dataset[k])
        plt.title(k)
    plt.show()
```

```python
graphique(dataset)
```
